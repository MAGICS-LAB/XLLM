```python
import matplotlib.pyplot as plt
import numpy as np
from palettable.lightbartlein.diverging import BlueDarkRed18_5

# Define grid parameters
num_rows = 15
num_cols = 63
square_size = 50
width = square_size * 0.65
height = square_size * 1
margin = 7

# Create figure and axes
fig, ax = plt.subplots(figsize=(30, 10))

# Define color range and colormap
vmin = 0
vmax = 3
cmap = BlueDarkRed18_5.mpl_colormap

# Loop through grid blocks
for i in range(num_rows):
   for j in range(num_cols):
       # Generate a random float value for the block
       value = np.random.rand() * 3

       # Calculate coordinates and color
       x = j * (width + margin)
       y = i * (height + margin)
       color = cmap(value / 3)

       # Draw the square
       rect = plt.Rectangle((x, y), width, height, color=color)
       ax.add_patch(rect)

# Set axis limits and turn off axes
ax.set_xlim(0, num_cols * (width + margin) + margin)
ax.set_ylim(0, num_rows * (height + margin) + margin)
ax.set_axis_off()

# Add colorbar
cax = fig.add_axes([0.95, 0.15, 0.02, 0.7])  # Position and size of colorbar
cbar = fig.colorbar(plt.cm.ScalarMappable(cmap=cmap, norm=plt.Normalize(vmin=vmin, vmax=vmax)), cax=cax)
cbar.set_label('Value Range')  # Set colorbar label

plt.show()

```
